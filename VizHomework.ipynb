```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'Top_Highest_Openings.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Opening'], df['Total Gross'], alpha=0.5)
plt.title('Opening vs Total Gross')
plt.xlabel('Opening Revenue in millions')
plt.ylabel('Total Gross Revenue in millions')
plt.grid(True)
plt.show()
```

```python
top10_movies = df.sort_values(by='Total Gross', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top10_movies['Release'], top10_movies['Total Gross'])
plt.title('Top 10 Movies by Total Gross')
plt.xlabel('Total Gross Revenue in millions')
plt.ylabel('Release')
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date')
total_gross_per_date = df.groupby('Date')['Total Gross'].sum()
plt.figure(figsize=(14, 7))
plt.plot(total_gross_per_date.index, total_gross_per_date.values)
plt.title('Total Gross Revenue Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Gross Revenue in millions')
plt.grid(True)
plt.show()
```
